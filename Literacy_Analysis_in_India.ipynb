Let us see the literacy rate wrt states

```python
plt.figure(figsize=(8, 10))
sns.barplot(x='OVERALL_LI', y='STATNAME', data=df1)
plt.xticks(rotation='vertical')
plt.xlabel('Literacy rate', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Overall Literacy Rate in India by States", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 10))
sns.barplot(x='FEMALE_LIT', y='STATNAME', data=df1)
plt.xticks(rotation='vertical')
plt.xlabel('Female Literacy rate', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Female Literacy Rate in India by States", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 10))
sns.barplot(x='MALE_LIT', y='STATNAME', data=df1)
plt.xticks(rotation='vertical')
plt.xlabel('Male Literacy rate', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Male Literacy Rate in India by States", fontsize=16)
plt.show()
```

Let us see the top 5 and bottom 5 states which have the highest women litercy rate

```python
top_5=df1.sort_values(by='OVERALL_LI', ascending=False).head(5)
bottom_5=df1.sort_values(by='OVERALL_LI', ascending=True).head(5)


top_bottom=pd.concat([top_5, bottom_5], axis=0, sort=False)
top_bottom
```

```python
plt.figure(figsize=(4, 6))
sns.barplot(x='OVERALL_LI', y='STATNAME', data=top_bottom, alpha=0.8)
plt.xticks(rotation='vertical')
plt.xlabel('Literacy rate', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Overall Literacy Rate in India - Top and Bottom States", fontsize=16)
plt.show()
```

Now let us see the which state has the highest difference in Male and Female Literacy

```python
top_bottom['Difference']=top_bottom['MALE_LIT']-top_bottom['FEMALE_LIT']
top_bottom['Difference']
```

```python
plt.figure(figsize=(4, 6))
sns.barplot(x='Difference', y='STATNAME', data=top_bottom, alpha=0.8)
plt.xticks(rotation='vertical')
plt.xlabel('Literacy rate difference', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Difference in Literacy Rate between Top and Bottom States in India", fontsize=16)
plt.show()
```

Let us find out the RURAL Population in India

```python
top_bottom['Rural']=100-top_bottom['P_URB_POP']
top_bottom['Rural']
```

```python
plt.figure(figsize=(10,8))

top_bottom.plot(y = ['P_URB_POP', 'Rural'], kind = 'bar')
plt.title("Comparison of Rural and Urban population")
```

Now let us see the sex ratio

```python
plt.figure(figsize=(12, 8))  # Updated figure size to (12, 8)
sns.barplot(x='STATNAME', y='SEXRATIO', data=top_bottom)
plt.xticks(rotation='vertical')  # Rotating x-axis labels vertically
plt.xlabel('States in India', fontsize=14)
plt.ylabel('Sex Ratio', fontsize=14)
plt.title("Sex Ratio in Different States of India", fontsize=16)
plt.show()
```
